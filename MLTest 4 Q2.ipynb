```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df=pd.read_csv(r'C:\Users\Dell\Desktop\Question-2\res\Retail.csv')
# print(data.info())
# print(data.Dept.value_counts())
df1 = df[(df['Dept'] !='0999:UNSCANNED ITEMS')]
# print(df1)
n0 = len(df1[(df1['Dept'] =='0973:CANDY')])
df1 = df1.fillna(0)
```

```python
df2 = df1.groupby(['POS Txn','Dept'])['Sales U'].sum().unstack().reset_index().fillna(0).set_index('POS Txn')
```

```python
def enc(x):
    if x <= 0:
        return 0
    elif x >=1:
        return 1
df3 = df2.applymap(enc)
```

```python
frq1 = apriori(df3, min_support=0.02, use_colnames=True)
sup1 = round(frq1.support.max(),5)
print(sup1)
```

```python
rules = association_rules(frq1, metric="lift", min_threshold=2)
n1 = len(rules)
rules
```

```python
f_rules = rules[(rules['lift']>=3)&(rules['confidence']>=0.1)]
n2 = len(f_rules)
f_rules
```

```python
result = [n0, sup1, n1, n2]
result
```

```python
result=pd.DataFrame(result)
result.to_csv('output.csv', header=False, index=False)
```
