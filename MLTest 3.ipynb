```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import confusion_matrix as c_m, classification_report as c_r, roc_auc_score as r_a_s, accuracy_score as a_s
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold, GridSearchCV
from sklearn.ensemble import AdaBoostClassifier
import sys
import sys
sys.path.append('../')

def solution():
    df=pd.read_csv('res/tic-tac-toe.data.txt',names=['x1','x2','x3','x4','x5','x6','x7','x8','x9','y'],sep=',')    

    df = df.replace('x',1)
    df = df.replace('o',0)
    df = df.replace('b',2)
    df = df.replace('positive',1)
    df = df.replace('negative',0)


    # print(df.y.value_counts())
    # print(df)

    X = df.drop('y',axis = 1)
    y = df.y
    X_train,X_test,y_train,y_test = tts(X,y,test_size = 0.25,random_state = 1, stratify = y)
    # print(X_train.shape)
    rfc_m1 = RandomForestClassifier(n_estimators = 100)
    rfc_m1.fit(X_train,y_train)
    y_pred1 = rfc_m1.predict(X_test)

    # print(c_r(y_test,y_pred1))
    k = 5
    kf = KFold(n_splits = k,shuffle = True,random_state = 0)
    acc_score = []
    for i,j in kf.split(X,y):
        X_train,X_test = X.iloc[i,:],X.iloc[j,:]
        y_train,y_test = y.iloc[i],y.iloc[j]

        rfc_m1.fit(X_train,y_train)
        y_pred2 = rfc_m1.predict(X_test)
        acc = a_s(y_test,y_pred2)
        acc_score.append(acc)
    avg_score1 = sum(acc_score)/k
    print('Accuracy_Before',avg_score1)
    
    abc_m2 = AdaBoostClassifier(base_estimator=RandomForestClassifier(), n_estimators=100, learning_rate=0.05, algorithm='SAMME.R', random_state=None)
    # abc_m2.fit(X_train,y_train)

    k = 20
    kf = KFold(n_splits = k,shuffle = True,random_state = 0)
    acc_score2 = []
    for i,j in kf.split(X,y):
        X_train,X_test = X.iloc[i,:],X.iloc[j,:]
        y_train,y_test = y.iloc[i],y.iloc[j]
        abc_m2.fit(X_train,y_train)
        y_pred3 = abc_m2.predict(X_test)
        acc2 = a_s(y_test,y_pred3)
        acc_score2.append(acc2)
    avg_score2 = sum(acc_score2)/k
    print('Accuracy_After',avg_score2)
    # list_acc = [avg_score1,avg_score2]
    # df2 = pd.DataFrame(list_acc,columns = ['AccuracyScores'])
    # df2.to_csv('output/output.csv')
    # # print(df2)

    result=[avg_score1,avg_score2]
    result=pd.DataFrame(result)
    #writing output to output.csv
    result.to_csv('output/output.csv', header=False, index=False)

solution()
```
