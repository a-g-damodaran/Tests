```python
basket = (df[df['Country'] == 'France'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack())
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    elif x >= 1:
        return 1
bsk_sets = bsk.applymap(encode_units)
bsk_sets.drop('POSTAGE', inplace = True, axis = 1)
bsk_sets
```

```python
print(bsk_sets['10 COLOUR SPACEBOY PEN'].value_counts())
```

```python
frequent_itemsets = apriori(bsk_sets,  min_support=0.07,use_colnames=True),
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
pdf1 = pd.read_excel(r'C:\Users\Dell\Desktop\DataFolder\OnlineRetail1.xlsx')
```

```python
pdf1 = og.copy()
```

```python
pdf1.dropna(axis = 0, subset = ['InvoiceNo'], inplace = True)
```

```python
pdf1['InvoiceNo'] = pdf1['InvoiceNo'].astype('str')
pdf1['Description']= pdf1['Description'].str.strip()
pdf1 = pdf1[~pdf1['InvoiceNo'].str.contains('C')]
```

```python
pdf1 = pdf1[pdf1['Country']=='France']
```

```python
pdf1
```

```python
pdf1 = pdf1.groupby(['InvoiceNo','Description']).Quantity.sum()
```

```python
pdf1 = pdf1.unstack()
```

```python
pdf1 = pdf1.fillna(0)
```

```python
pdf1
```

```python
def encod1(x):
    if x <= 0:
        return 0
    elif x >=1:
        return 1
```

```python
bsk1 = pdf1.applymap(encod1)
```

```python
bsk1.drop('POSTAGE', axis = 1, inplace = True)
```

```python
bsk1
```

```python
frq1 = apriori(bsk1,min_support=0.1,use_colnames=True,max_len=None,verbose=0,low_memory=False,)
```

```python
frq1
```

```python
frq2 = association_rules(frq1, metric='lift', min_threshold=1,support_only=False,)
```

```python
frq2
```

```python
frq3 = association_rules(frq1, metric='confidence', min_threshold=0.7,support_only=False,)
```
