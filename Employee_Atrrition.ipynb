```python
df1=df.copy()
# data['Order Date']=pd.to_datetime(data['Order Date'],format='%d-%m-%Y')
df1['Dateofjoining'] = pd.to_datetime(df1['Dateofjoining'], format= '%d-%m-%Y')
df1['DOJ_Y']=df1['Dateofjoining'].dt.year
df1['DOJ_M']=df1['Dateofjoining'].dt.month
```

```python
df1['LastWorkingDate'] = pd.to_datetime(df1['LastWorkingDate'], format= '%d-%m-%Y')
df1['LW_Y']=df1['LastWorkingDate'].dt.year
df1['LW_M']=df1['LastWorkingDate'].dt.month
```

```python
df1['LW_Y']=df1['LW_Y'].astype('str')
df1['LW_Y'].replace('nan','0',inplace = True)
df1['LW_M']=df1['LW_M'].astype('str')
df1['LW_M'].replace('nan','0',inplace = True)
```

```python
df1.drop(['Dateofjoining','LastWorkingDate'], axis = 1,inplace = True)
```

```python
df1['LW_Y']=df1['LW_Y'].astype('float')
df1['LW_M']=df1['LW_M'].astype('float')
```

```python
df1.Gender.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['Gender'] = le.fit_transform(df1['Gender'])
```

```python
df_dum = pd.get_dummies(df1[['City','Education_Level']])
```

```python
df1 = pd.concat([df1,df_dum], axis = 1)
```

```python
df1.head()
```

```python
df2 = df1.groupby(by='Emp_ID').sum().reset_index()
```

```python
df2['Status'] = 0
df2
```

```python
for i in range(len(df2)):
  if (df2.LW_Y[i] == 0)	& (df2.LW_M[i] == 0) :
    status = 0
  else:
    status = 1

  df2['Status'][i] = status
```

```python
df2
```

```python
df3 = df2[['LW_Y',	'LW_M', 'Status']]
```

```python
# df3['LW_Y']=df1['LW_Y'].astype('str')
df3['LW_Y'].replace(0.0,2017,inplace = True)
# df3['LW_M']=df1['LW_M'].astype('str')
df3['LW_M'].replace(0.0,12,inplace = True)
```

```python
df_f
```

```python
df_f['Duration'] = 0
```

```python
for i in range(len(df_f)):
  duration = ((df_f.LW_Y[i] - 1)  - (df_f.DOJ_Y[i]))*2
  if df_f.DOJ_M[i] <= 6 :
    duration = duration + 2
  else:
    duration = duration + 1
  if df_f.LW_M[i] >= 6:
    duration = duration + 2
  else:
    duration = duration + 1
  df_f['Duration'][i] = duration
```
